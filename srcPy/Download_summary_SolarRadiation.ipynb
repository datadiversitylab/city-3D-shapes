```python
# Solar radiation Mean FY 2018
worldGeometry = ee.Geometry.Polygon([
  [-180, 90],  #// Top-left corner
  [180, 90],   #// Top-right corner
  [180, -90],  #// Bottom-right corner
  [-180, -90]  #// Bottom-left corner
])
cty = ee.FeatureCollection('projects/ee-pradnyauoa/assets/data')

out_dir = os.path.expanduser("~/Downloads")
loc = worldGeometry   #fc1 # worldGeometry #cty.geometry() #ee.Geometry.Point(-99.2222, 46.7816)

collection = ee.ImageCollection('MODIS/061/MCD18A1') \
                  .filter(ee.Filter.date('2018-01-01', '2018-12-31')) \
                  .select(['GMT_0000_DSR', 'GMT_0300_DSR', 'GMT_0600_DSR', 'GMT_0900_DSR', 'GMT_1200_DSR', 'GMT_1500_DSR', 'GMT_1800_DSR', 'GMT_2100_DSR' ]).mean()

geemap.download_ee_image(collection, "Solar_Radiation_2018Mean_30m_scale30.tif", region = worldGeometry, scale=30, crs='EPSG:4326')

```

```python
gmt0000 = collection.select(['GMT_0000_DSR' ])
gmt0300 = collection.select(['GMT_0300_DSR' ])
gmt0600 = collection.select(['GMT_0600_DSR' ])
gmt0900 = collection.select(['GMT_0900_DSR' ])
gmt1200 = collection.select(['GMT_1200_DSR' ])
gmt1500 = collection.select(['GMT_1500_DSR' ])
gmt1800 = collection.select(['GMT_1800_DSR' ])
gmt2100 = collection.select(['GMT_2100_DSR' ])

collBandMean = ee.Image.cat([gmt0000 , gmt0300, gmt0600, gmt0900, gmt1200, gmt1500, gmt1800, gmt2100 ]).reduce(ee.Reducer.mean());
geemap.download_ee_image(collBandMean, "Solar_Radiation_2018BandMean_30m_scale30.tif", region = worldGeometry, scale=30, crs='EPSG:4326')
```
